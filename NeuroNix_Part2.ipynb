```python
import os
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split

# Define preprocessing pipeline for images
preprocessing_steps = transforms.Compose([
    transforms.Resize((128, 128)),  # Standardize image dimensions
    transforms.ToTensor(),
    transforms.Normalize(mean=(0.5,), std=(0.5,))  # Normalize assuming grayscale images
])

# Root directory containing dataset (update as necessary)
data_directory = "/content/drive/MyDrive/Colab_Notebooks/Datasets/rice_leaf_diseases"

# Load the dataset using ImageFolder with defined transformations
image_dataset = datasets.ImageFolder(root=data_directory, transform=preprocessing_steps)

# Determine dataset sizes for training, validation, and testing
dataset_size = len(image_dataset)
train_split = int(0.6 * dataset_size)
validation_split = int(0.2 * dataset_size)
test_split = dataset_size - train_split - validation_split

# Divide dataset into subsets for training, validation, and testing
training_subset, validation_subset, testing_subset = random_split(
    image_dataset, [train_split, validation_split, test_split]
)

# DataLoader configuration
batch_size = 64  # Number of samples per batch
training_loader = DataLoader(training_subset, batch_size=batch_size, shuffle=True, num_workers=2)
validation_loader = DataLoader(validation_subset, batch_size=batch_size, shuffle=False, num_workers=2)
testing_loader = DataLoader(testing_subset, batch_size=batch_size, shuffle=False, num_workers=2)

# Display the mapping of classes to indices
print("Category to Index Mapping:", image_dataset.class_to_idx)

# Display counts for each data split
print(f"Training Samples: {len(training_subset)}")
print(f"Validation Samples: {len(validation_subset)}")
print(f"Testing Samples: {len(testing_subset)}")
```

```python
import torch
import torchvision
from torchvision import transforms, models
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
import matplotlib.pyplot as plt

# Function to load and customize a pre-trained model for Rice Leaf Disease classification
def initialize_model(model_name):
    if model_name == 'resnet':
        base_model = models.resnet18(pretrained=True)
        input_features = base_model.fc.in_features
        base_model.fc = nn.Linear(input_features, 3)  # Modify the classifier
    elif model_name == 'densenet':
        base_model = models.densenet121(pretrained=True)
        input_features = base_model.classifier.in_features
        base_model.classifier = nn.Linear(input_features, 3)  # Modify the classifier
    elif model_name == 'googlenet':
        base_model = models.googlenet(pretrained=True)
        input_features = base_model.fc.in_features
        base_model.fc = nn.Linear(input_features, 3)  # Modify the classifier
    else:
        raise ValueError("Unsupported model name provided. Use 'resnet', 'densenet', or 'googlenet'.")

    return base_model

# Function for model training and validation
def train_and_evaluate(model, train_loader, validation_loader, epochs=20, lr=1e-5):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = model.to(device)

    loss_function = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    # Track losses
    training_losses, validation_losses = [], []

    # Training loop
    for epoch in range(epochs):
        model.train()
        epoch_training_loss = 0.0
        for inputs, targets in train_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            optimizer.zero_grad()
            predictions = model(inputs)
            loss = loss_function(predictions, targets)
            loss.backward()
            optimizer.step()
            epoch_training_loss += loss.item()

        training_losses.append(epoch_training_loss / len(train_loader))

        # Validation step
        model.eval()
        epoch_validation_loss = 0.0
        with torch.no_grad():
            for inputs, targets in validation_loader:
                inputs, targets = inputs.to(device), targets.to(device)
                predictions = model(inputs)
                loss = loss_function(predictions, targets)
                epoch_validation_loss += loss.item()

        validation_losses.append(epoch_validation_loss / len(validation_loader))

        print(f"Epoch {epoch + 1}/{epochs} - Train Loss: {training_losses[-1]:.4f}, Validation Loss: {validation_losses[-1]:.4f}")

    return training_losses, validation_losses

# Function to plot training and validation losses
def plot_losses(train_losses, val_losses, title):
    plt.plot(range(1, len(train_losses) + 1), train_losses, label='Training Loss')
    plt.plot(range(1, len(val_losses) + 1), val_losses, label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title(title)
    plt.legend()
    plt.show()

# Number of training epochs
num_epochs = 20

# Model names to experiment with
model_variants = ['resnet', 'densenet', 'googlenet']

# Dictionary to store fine-tuned models
fine_tuned_models = {}

# Training and evaluation loop for each model
for model_name in model_variants:
    print(f"\nFine-tuning {model_name.upper()} for Rice Leaf Disease Classification:")
    selected_model = initialize_model(model_name)
    train_losses, val_losses = train_and_evaluate(selected_model, training_loader, testing_loader, num_epochs, lr=0.0001)

    # Save the fine-tuned model for later use
    fine_tuned_models[model_name] = selected_model

    # Final loss display
    print(f"\nFinal {model_name.upper()} Training Loss: {train_losses[-1]:.4f}")
    print(f"Final {model_name.upper()} Validation Loss: {val_losses[-1]:.4f}")

    # Plot the loss curves
    plot_losses(train_losses, val_losses, f"Training and Validation Loss ({model_name.upper()})")
```

```python
import torch
from sklearn.metrics import accuracy_score

# Function to assess model performance on the test dataset
def assess_model_performance(model, test_loader):
    """
    Evaluate the given model using the test data loader. Test accuracy as a percentage.
    """
    model.eval()  # Set the model to evaluation mode
    predictions_list = []
    true_labels_list = []

    device = next(model.parameters()).device  # Retrieve the model's device (CPU/GPU)

    with torch.no_grad():
        for batch_inputs, batch_labels in test_loader:
            # Transfer inputs and labels to the same device as the model
            batch_inputs, batch_labels = batch_inputs.to(device), batch_labels.to(device)

            # Perform forward pass
            batch_outputs = model(batch_inputs)

            # Get predicted class indices
            _, predicted_classes = torch.max(batch_outputs, dim=1)

            # Accumulate predictions and true labels
            predictions_list.extend(predicted_classes.cpu().numpy())
            true_labels_list.extend(batch_labels.cpu().numpy())

    # Compute test accuracy
    test_accuracy = accuracy_score(true_labels_list, predictions_list) * 100
    return test_accuracy

# Load fine-tuned models from the fine-tuning step
pre_trained_resnet_model = fine_tuned_models['resnet']
pre_trained_densenet_model = fine_tuned_models['densenet']
pre_trained_googlenet_model = fine_tuned_models['googlenet']

# Evaluate the fine-tuned models
resnet_test_accuracy = assess_model_performance(pre_trained_resnet_model, testing_loader)
densenet_test_accuracy = assess_model_performance(pre_trained_densenet_model, testing_loader)
googlenet_test_accuracy = assess_model_performance(pre_trained_googlenet_model, testing_loader)

# Display the test accuracies
print(f"ResNet Test Accuracy: {resnet_test_accuracy:.2f}%")
print(f"DenseNet Test Accuracy: {densenet_test_accuracy:.2f}%")
print(f"GoogLeNet Test Accuracy: {googlenet_test_accuracy:.2f}%")
```
